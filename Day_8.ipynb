```python
import pandas as pd
data={
    "region":["north","south","west",],
    "product":["laptop","mobile","tab"],
    "sales":[25000,15000,76000]
    }
df=pd.DataFrame(data)
print(df)

```

```python
print(df['sales'].sum())
```

```python
print(df['sales'].max())
```

```python
print(df['sales'].min())
```
