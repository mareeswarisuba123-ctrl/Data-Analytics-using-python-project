```python
Mark1=int(input("Enter M1:"))
Mark2=int(input("Enter M2:"))
Mark3=int(input("Enter M3:"))

total = Mark1 + Mark2 + Mark3
Average=total/3

print("total:",total)
print("Average:",Average)

if Average >= 90:
    print("Grade: A")
elif Average >= 50:
    print("Grade: B")
else:
    print("Grade: C")
```
